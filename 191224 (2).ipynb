```python
plt.figure(figsize=(20,7))
ax= sns.countplot(x='Age',data=df)
ax.bar_label(ax.containers[0]) 
plt.title('Age Disttribution')
plt.show()
```

**Most of the peoples are of 69 age**

```python
ax = sns.countplot(x='Gender', data=df)
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```


**The number of males are more than the females**

```python
#2
ax = sns.countplot(hue='Gender',x='Subscription Status', data=df)
ax.bar_label(ax.containers[0])
plt.title('Gender wise Subscription Status Distribution')
plt.show()
```

```python
plt.figure(figsize=(20,5))  
ax = sns.countplot(x='Item Purchased', data=df)  
ax.bar_label(ax.containers[0])  
plt.show()
```


**Most sales are of Blouse, Pants, and Jewelry.**

```python
plt.figure(figsize=(10,5))  
ax = sns.countplot(x='Category', data=df)  
ax.bar_label(ax.containers[0]) 
plt.show()
```


**Clothing category is most among all others.**

```python
#2
sns.catplot(hue='Category',kind='count',x='Gender',data=df)
plt.title('Gender by Category')
plt.show()
```

```python
plt.figure(figsize=(20,5))  
ax = sns.countplot(x='Location', data=df)  
plt.xticks(rotation=90)  
ax.bar_label(ax.containers[0])  
plt.show()
```

**Most buyers are from Montana, California, and Idaho.**

```python
plt.figure(figsize=(10,5))  
ax = sns.countplot(x='Size', data=df)  
ax.bar_label(ax.containers[0]) 
plt.show()
```

**Medium size products have the highest range of sell.**

```python
#2
sns.catplot(hue='Category',kind='count',x='Size',data=df)
plt.title('Gender by Category')
plt.show()
```

```python
plt.figure(figsize=(20,5))  
ax = sns.countplot(x='Color', data=df)  
plt.xticks(rotation=90)  
ax.bar_label(ax.containers[0])  
plt.show()
```

**Buyers bought Olive, Yellow, and Silver colors most.**

```python
plt.figure(figsize=(10,5)) 
ax = sns.countplot(x='Season', data=df)  
ax.bar_label(ax.containers[0]) 
plt.show()
```

**In Spring, Fall, Winter and then Summer seasons the sell out is most.**

```python
sales = df.groupby('Gender')['Purchase Amount (USD)'].sum().reset_index().sort_values(by='Purchase Amount (USD)', ascending=False)
sns.barplot(data=sales, x='Gender', y='Purchase Amount (USD)')
plt.show()
```

```python
plt.figure(figsize=(20,5))
sales = df.groupby(['Item Purchased'],as_index=False)['Purchase Amount (USD)'].sum().sort_values(by='Purchase Amount (USD)',ascending=False)
sns.barplot(data=sales,x='Item Purchased',y='Purchase Amount (USD)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sales = df.groupby(['Category'],as_index=False)['Purchase Amount (USD)'].sum().sort_values(by='Purchase Amount (USD)',ascending=False)
sns.barplot(data=sales,x='Category',y='Purchase Amount (USD)')
plt.show()
```

**Clothing category has the most Purchased amount.**


```python
plt.figure(figsize=(20,5))
sales = df.groupby(['Location'],as_index=False)['Purchase Amount (USD)'].sum().sort_values(by='Purchase Amount (USD)',ascending=False)
sns.barplot(data=sales,x='Location',y='Purchase Amount (USD)')
plt.xticks(rotation=90)
plt.show()
```

**Montana has peoduced the most amount.**
